````python
import os
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset
import torchaudio
from transformers import AutoProcessor
import warnings
warnings.filterwarnings('ignore')

BATCH_SIZE = 16
EPOCHS = 10
LEARNING_RATE = 1e-4
MAX_AUDIO_LENGTH = 30  # seconds

def prepare_dataset(batch, processor):
    """Process audio and prepare labels"""
    # Process the audio
    audio_arrays = [audio['array'] for audio in batch['audio']]
    processed_audio = processor(
        audio=audio_arrays,
        sampling_rate=processor.sampling_rate,
        return_tensors="pt",
        padding=True
    )

    # Convert genre strings to numeric labels
    genre_to_id = {
        'Electronic': 0, 'Experimental': 1, 'Folk': 2, 'Hip-Hop': 3,
        'Instrumental': 4, 'International': 5, 'Pop': 6, 'Rock': 7
    }
    labels = [genre_to_id[genre] for genre in batch['genre']]

    return {
        'input_values': processed_audio.input_values,
        'attention_mask': processed_audio.attention_mask,
        'labels': labels
    }
````

````python
def create_dataloaders(dataset, processor, batch_size):
    """Create training and validation dataloaders"""
    # Split dataset into train and validation
    dataset = dataset['train'].train_test_split(test_size=0.2, shuffle=True, seed=42)

    # Process datasets
    train_dataset = dataset['train'].map(
        lambda x: prepare_dataset(x, processor),
        remove_columns=dataset['train'].column_names,
        batch_size=batch_size,
        batched=True,
        num_proc=4
    )

    val_dataset = dataset['test'].map(
        lambda x: prepare_dataset(x, processor),
        remove_columns=dataset['test'].column_names,
        batch_size=batch_size,
        batched=True,
        num_proc=4
    )

    # Set format for pytorch
    train_dataset.set_format(type='torch', columns=['input_values', 'attention_mask', 'labels'])
    val_dataset.set_format(type='torch', columns=['input_values', 'attention_mask', 'labels'])

    # Create dataloaders
    train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_dataloader = DataLoader(val_dataset, batch_size=batch_size)

    return train_dataloader, val_dataloader
````

````python
from functools import lru_cache
def load_metadata(data_path):
    """Load and process FMA metadata."""
    tracks = pd.read_csv(os.path.join(data_path, 'fma_metadata/tracks_clean.csv'),
                        index_col=0, header=[0, 1])

    # Select relevant columns
    keep_cols = [('track', 'genre_top'), ('track', 'title'),
                ('track', 'duration')]
    tracks = tracks[keep_cols]

    # Clean up data
    tracks = tracks[tracks[('track', 'genre_top')].notna()]
    tracks = tracks[tracks[('track', 'duration')].notna()]

    # Create genre mapping
    genres = sorted(tracks[('track', 'genre_top')].unique())
    genre_to_id = {genre: i for i, genre in enumerate(genres)}

    return tracks, genre_to_id

def get_audio_path(track_id, data_path):
    """Get the path of an audio file from its track ID."""
    tid_str = '{:06d}'.format(track_id)
    return os.path.join(data_path, 'fma_small',
                       tid_str[:3], tid_str + '.mp3')

import os
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
import torchaudio
from transformers import ASTFeatureExtractor
from functools import lru_cache

class FMADataset(Dataset):
    def __init__(self, data_path, max_length=(16_000*30)):  # ~10 seconds at 16kHz
        self.data_path = data_path
        self.max_length = max_length

        # Load metadata
        metadata_path = os.path.join(data_path, 'fma_metadata/tracks_clean.csv')
        self.tracks = pd.read_csv(metadata_path, index_col=0, header=[0, 1])
        self.track_ids = list(self.tracks.index)

        # Create genre mapping
        genres = sorted(self.tracks[('track', 'genre_top')].unique())
        self.genre_to_id = {genre: i for i, genre in enumerate(genres)}

        # Initialize feature extractor
        self.feature_extractor = ASTFeatureExtractor()

        # Initialize resampler
        self.resampler = torchaudio.transforms.Resample(
            orig_freq=44100,  # Original FMA sample rate
            new_freq=16000    # Target sample rate for AST
        )

        print("Genre mapping:", self.genre_to_id)

    @lru_cache(maxsize=None)
    def get_audio_path(self, track_id):
        """Cached function to get audio file path."""
        tid_str = '{:06d}'.format(track_id)
        return os.path.join(self.data_path, 'fma_small',
                           tid_str[:3], tid_str + '.mp3')

    def __len__(self):
        return len(self.track_ids)

    def __getitem__(self, idx):
        track_id = self.track_ids[idx]
        audio_path = self.get_audio_path(track_id)

        try:
            # Load audio
            waveform, original_sr = torchaudio.load(audio_path)

            # Convert to mono if stereo
            if waveform.size(0) > 1:
                waveform = waveform.mean(dim=0, keepdim=True)

            # Resample to 16kHz
            waveform = self.resampler(waveform)

            # Convert to numpy and limit length if necessary
            waveform = waveform.squeeze().numpy()
            # if len(waveform) > self.max_length:
            #     waveform = waveform[:self.max_length]

            # Extract features using AST feature extractor
            inputs = self.feature_extractor(
                waveform,
                sampling_rate=16000,  # Now we know it's 16kHz
                padding="max_length",
                return_tensors="pt"
            )

            # Get features and squeeze batch dimension added by feature_extractor
            features = inputs.input_values.squeeze(0)

            # Get label
            genre = self.tracks.loc[track_id, ('track', 'genre_top')]
            label = self.genre_to_id[genre]

            return features, label

        except Exception as e:
            print(f"Error loading {audio_path}: {e}")
            # Return zero tensor with correct shape (1024, 128) and label
            return torch.zeros((1024, 128)), 0

def create_dataloaders(dataset, batch_size=8, train_ratio=0.8):
    """Create train and validation dataloaders."""
    # Calculate lengths
    train_size = int(train_ratio * len(dataset))
    val_size = len(dataset) - train_size

    # Split dataset
    train_dataset, val_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size]
    )

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True
    )

    return train_loader, val_loader
````

````python
def create_data_loaders(
    dataset,
    batch_size=16,  # Reduced batch size
    train_split=0.8,
    num_workers=4,
    pin_memory=True,
    prefetch_factor=2
):
    """Create optimized data loaders."""
    # Create train/validation split
    train_size = int(train_split * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size]
    )

    # Create data loaders with optimized settings
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=num_workers,
        pin_memory=pin_memory,
        prefetch_factor=prefetch_factor,
        persistent_workers=True,
        drop_last=True  # Helps with batch norm and slightly improves speed
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=num_workers,
        pin_memory=pin_memory,
        prefetch_factor=prefetch_factor,
        persistent_workers=True,
        drop_last=False  # Keep all validation samples
    )

    return train_loader, val_loader
````

````python
import torch
from transformers import ASTForAudioClassification, AutoFeatureExtractor
from datasets import load_dataset
from torch.utils.data import DataLoader
import torch.cuda.amp as amp
from tqdm import tqdm

class AudioDataset(torch.utils.data.Dataset):
    def __init__(self, hf_dataset, feature_extractor):
        self.dataset = hf_dataset
        self.feature_extractor = feature_extractor

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]

        # Get audio array and sampling rate
        audio = item['audio']

        # Process audio with feature extractor
        inputs = self.feature_extractor(
            audio['array'],
            sampling_rate=audio['sampling_rate'],
            padding="max_length",
            return_tensors="pt"
        )

        # Get features and label
        features = inputs.input_values.squeeze(0)
        label = item['genre']

        return features, label

def create_dataloaders(dataset_name, batch_size=32):
    """Create dataloaders from HuggingFace dataset."""
    # Load dataset
    dataset = load_dataset(dataset_name)

    # Load feature extractor
    feature_extractor = AutoFeatureExtractor.from_pretrained(
        "MIT/ast-finetuned-audioset-10-10-0.4593"
    )

    # Create train and validation datasets
    train_dataset = AudioDataset(dataset['train'], feature_extractor)
    val_dataset = AudioDataset(dataset['validation'], feature_extractor)

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True
    )

    return train_loader, val_loader
````

````python
from dataclasses import dataclass
import torch
from torch.optim import AdamW, Adam, SGD
from torch.optim.lr_scheduler import OneCycleLR, CosineAnnealingLR, LinearLR

@dataclass
class TrainingConfig:
    # Core training parameters
    batch_size: int = 32
    num_epochs: int = 50
    learning_rate: float = 1e-4
    weight_decay: float = 0.01

    # Model/Data parameters
    max_audio_length: int = 160000  # 10 seconds at 16kHz

    # Optimizer settings
    optimizer_type: str = 'adamw'  # 'adamw', 'adam', or 'sgd'
    scheduler_type: str = 'onecycle'  # 'onecycle', 'cosine', or 'linear'

    # Training techniques
    gradient_clip_val: float = 1.0
    label_smoothing: float = 0.1
    warmup_ratio: float = 0.1  # Percentage of total steps for warmup

    # Optimizer specific parameters
    momentum: float = 0.9  # For SGD
    beta1: float = 0.9    # For Adam/AdamW
    beta2: float = 0.999  # For Adam/AdamW

def get_optimizer(model, config: TrainingConfig):
    """Get optimizer based on config."""
    if config.optimizer_type == 'adamw':
        return AdamW(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            betas=(config.beta1, config.beta2)
        )
    elif config.optimizer_type == 'adam':
        return Adam(
            model.parameters(),
            lr=config.learning_rate,
            betas=(config.beta1, config.beta2)
        )
    elif config.optimizer_type == 'sgd':
        return SGD(
            model.parameters(),
            lr=config.learning_rate,
            momentum=config.momentum,
            weight_decay=config.weight_decay
        )
    else:
        raise ValueError(f"Unknown optimizer type: {config.optimizer_type}")

def get_scheduler(optimizer, num_training_steps, config: TrainingConfig):
    """Get learning rate scheduler based on config."""
    if config.scheduler_type == 'onecycle':
        return OneCycleLR(
            optimizer,
            max_lr=config.learning_rate,
            total_steps=num_training_steps,
            pct_start=config.warmup_ratio
        )
    elif config.scheduler_type == 'cosine':
        return CosineAnnealingLR(
            optimizer,
            T_max=num_training_steps
        )
    elif config.scheduler_type == 'linear':
        return LinearLR(
            optimizer,
            start_factor=1.0,
            end_factor=0.01,
            total_iters=num_training_steps
        )
    else:
        raise ValueError(f"Unknown scheduler type: {config.scheduler_type}")

# Example configurations for different scenarios
default_config = TrainingConfig(num_epochs=10)

fast_training_config = TrainingConfig(
    batch_size=64,
    num_epochs=3,
    learning_rate=2e-4,
    scheduler_type='linear',
    warmup_ratio=0.05
)

conservative_config = TrainingConfig(
    batch_size=16,
    learning_rate=5e-5,
    num_epochs=3,
    weight_decay=0.1,
    gradient_clip_val=0.5,
    optimizer_type='adamw',
    scheduler_type='cosine'
)

large_batch_config = TrainingConfig(
    batch_size=128,
    learning_rate=4e-4,  # Scaled up for larger batch
    num_epochs=3,
    optimizer_type='adamw',
    scheduler_type='onecycle',
    warmup_ratio=0.2
)
````

````python
import wandb
import torch
import torch.nn as nn
from torch.cuda.amp import autocast, GradScaler
from tqdm import tqdm
from dataclasses import asdict

def train_model(model, train_loader, val_loader, config: TrainingConfig, project_name="ast-genre-classification", device='cuda'):
    """
    Train the model with WandB logging.
    """
    # Initialize WandB
    wandb.init(
        project=project_name,
        config=asdict(config),  # Log all config parameters
        name=f"{config.optimizer_type}_bs{config.batch_size}_lr{config.learning_rate}"
    )

    # Watch the model to log gradients and parameters
    wandb.watch(model, log="all", log_freq=10)

    # Calculate total training steps
    num_training_steps = len(train_loader) * config.num_epochs

    # Setup
    optimizer = get_optimizer(model, config)
    scheduler = get_scheduler(optimizer, num_training_steps, config)
    scaler = GradScaler()
    criterion = nn.CrossEntropyLoss(label_smoothing=config.label_smoothing)

    best_val_acc = 0
    step = 0

    for epoch in range(config.num_epochs):
        # Training
        model.train()
        train_loss = 0
        pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{config.num_epochs}')

        for batch_idx, (features, labels) in enumerate(pbar):
            features, labels = features.to(device), labels.to(device)

            with autocast():
                outputs = model(input_values=features, labels=labels)
                loss = outputs.loss

            scaler.scale(loss).backward()

            if config.gradient_clip_val > 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(
                    model.parameters(),
                    config.gradient_clip_val
                )

            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            scheduler.step()

            train_loss += loss.item()

            # Log training metrics
            if batch_idx % 10 == 0:
                wandb.log({
                    "train/loss": loss.item(),
                    "train/learning_rate": scheduler.get_last_lr()[0],
                    "train/step": step,
                    "train/epoch": epoch + batch_idx/len(train_loader)
                })

            step += 1
            pbar.set_postfix({'loss': loss.item()})

        avg_train_loss = train_loss / len(train_loader)

        # Validation
        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for features, labels in tqdm(val_loader, desc='Validation'):
                features, labels = features.to(device), labels.to(device)
                outputs = model(input_values=features, labels=labels)
                loss = outputs.loss
                logits = outputs.logits

                val_loss += loss.item()
                _, predicted = logits.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

                # Store predictions and labels for confusion matrix
                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        avg_val_loss = val_loss / len(val_loader)
        val_accuracy = correct / total

        # Create confusion matrix
        wandb.log({
            "val/conf_mat": wandb.plot.confusion_matrix(
                probs=None,
                y_true=all_labels,
                preds=all_preds,
                class_names=list(range(8))  # Assuming 8 genres
            )
        })

        # Log validation metrics
        wandb.log({
            "val/loss": avg_val_loss,
            "val/accuracy": val_accuracy,
            "train/epoch_loss": avg_train_loss,
            "epoch": epoch
        })

        print(f"\nEpoch {epoch+1} Results:")
        print(f"Train Loss: {avg_train_loss:.4f}")
        print(f"Val Loss: {avg_val_loss:.4f}")
        print(f"Val Accuracy: {val_accuracy:.4f}")

        # Save best model
        if val_accuracy > best_val_acc:
            best_val_acc = val_accuracy
            model_path = f'best_model_{config.optimizer_type}_{config.batch_size}.pth'
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'config': config,
                'val_accuracy': val_accuracy,
            }, model_path)

            # Log best model artifact to wandb
            artifact = wandb.Artifact(
                name=f"model-{wandb.run.id}",
                type="model",
                description=f"Best model with accuracy {val_accuracy:.4f}"
            )
            artifact.add_file(model_path)
            wandb.log_artifact(artifact)

            print(f"Saved new best model with accuracy: {val_accuracy:.4f}")

    wandb.finish()
````

````python
import wandb
import torch
from transformers import ASTForAudioClassification
from tqdm import tqdm
import torch.cuda.amp as amp
from dataclasses import asdict

def train_model(model, train_loader, val_loader, num_epochs=10, device='cuda', class_names=None):
    """
    Train the AST model with comprehensive W&B logging.

    Args:
        model: The model to train
        train_loader: Training data loader
        val_loader: Validation data loader
        num_epochs: Number of training epochs
        device: Device to train on
        class_names: List of class names for confusion matrix
    """
    # Initialize wandb with hardcoded config
    wandb.init(
        project="ast-audio-classification",  # You can change this project name
        config={
            "learning_rate": 1e-4,
            "batch_size": train_loader.batch_size,
            "num_epochs": num_epochs,
            "optimizer": "AdamW",
            "scheduler": "OneCycleLR"
        },
        name=f"AST_bs{train_loader.batch_size}_lr1e-4"
    )

    # Watch model for gradient and parameter logging
    wandb.watch(model, log="all", log_freq=10)

    # Setup optimization
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=1e-4,
        epochs=num_epochs,
        steps_per_epoch=len(train_loader)
    )
    scaler = amp.GradScaler()

    # Initialize best validation accuracy
    best_val_acc = 0
    step = 0

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch + 1}/{num_epochs}")

        # Training phase
        model.train()
        train_loss = 0
        train_pbar = tqdm(train_loader, desc='Training')

        train_predictions = []
        train_ground_truth = []

        for features, labels in train_pbar:
            features, labels = features.to(device), labels.to(device)

            # Mixed precision training
            with amp.autocast():
                outputs = model(input_values=features, labels=labels)
                loss = outputs.loss

            # Backward pass
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            scheduler.step()

            # Get predictions for confusion matrix
            logits = outputs.logits
            _, predicted = logits.max(1)
            train_predictions.extend(predicted.cpu().numpy())
            train_ground_truth.extend(labels.cpu().numpy())

            # Log training metrics
            wandb.log({
                "train/step_loss": loss.item(),
                "train/learning_rate": scheduler.get_last_lr()[0],
                "train/step": step
            })

            train_loss += loss.item()
            train_pbar.set_postfix({'loss': loss.item()})
            step += 1

        avg_train_loss = train_loss / len(train_loader)

        # Log training confusion matrix
        wandb.log({
            "train/confusion_matrix": wandb.plot.confusion_matrix(
                probs=None,
                y_true=train_ground_truth,
                preds=train_predictions,
                class_names=class_names
            )
        })

        # Validation phase
        model.eval()
        val_loss = 0
        correct = 0
        total = 0

        val_predictions = []
        val_ground_truth = []
        val_probabilities = []

        with torch.no_grad():
            val_pbar = tqdm(val_loader, desc='Validation')
            for features, labels in val_pbar:
                features, labels = features.to(device), labels.to(device)

                outputs = model(input_values=features, labels=labels)
                loss = outputs.loss
                logits = outputs.logits

                # Get predictions and probabilities
                probabilities = torch.nn.functional.softmax(logits, dim=1)
                _, predicted = logits.max(1)

                val_predictions.extend(predicted.cpu().numpy())
                val_ground_truth.extend(labels.cpu().numpy())
                val_probabilities.extend(probabilities.cpu().numpy())

                val_loss += loss.item()
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

                val_pbar.set_postfix({
                    'loss': loss.item(),
                    'accuracy': 100 * correct / total
                })

        # Calculate metrics
        avg_val_loss = val_loss / len(val_loader)
        val_accuracy = correct / total

        # Log validation metrics
        wandb.log({
            "train/epoch_loss": avg_train_loss,
            "val/epoch_loss": avg_val_loss,
            "val/accuracy": val_accuracy,
            "epoch": epoch,

            # Confusion matrix for validation set
            "val/confusion_matrix": wandb.plot.confusion_matrix(
                probs=None,
                y_true=val_ground_truth,
                preds=val_predictions,
                class_names=class_names
            ),

            # ROC curve
            "val/roc_curve": wandb.plot.roc_curve(
                y_true=val_ground_truth,
                y_probas=val_probabilities,
                labels=class_names
            ),

            # PR curve
            "val/pr_curve": wandb.plot.pr_curve(
                y_true=val_ground_truth,
                y_probas=val_probabilities,
                labels=class_names
            )
        })

        print(f"Train Loss: {avg_train_loss:.4f}")
        print(f"Val Loss: {avg_val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")

        # Save best model
        if val_accuracy > best_val_acc:
            best_val_acc = val_accuracy
            torch.save(model.state_dict(), 'best_ast_model.pth')

            # Log best model to W&B
            model_artifact = wandb.Artifact(
                name=f"best_model_epoch_{epoch}",
                type="model",
                description=f"Best model with validation accuracy: {val_accuracy:.4f}"
            )
            model_artifact.add_file('best_ast_model.pth')
            wandb.log_artifact(model_artifact)

            print(f"Saved new best model with accuracy: {val_accuracy:.4f}")

    # Close wandb run
    wandb.finish()
````

````python
import os
import pandas as pd
import numpy as np
from datasets import Dataset, DatasetDict, Audio, Features, Value, ClassLabel
from huggingface_hub import HfApi
import torchaudio
import json

def prepare_fma_dataset(data_dir):
    """Prepare FMA dataset for Hugging Face upload."""

    # Load and process metadata
    metadata_path = os.path.join(data_dir, 'fma_metadata/tracks_clean.csv')
    tracks = pd.read_csv(metadata_path, index_col=0, header=[0, 1])

    # Create genre mapping
    genres = sorted(tracks[('track', 'genre_top')].unique())
    genre_to_id = {genre: i for i, genre in enumerate(genres)}

    # Save genre mapping for later use
    with open('genre_mapping.json', 'w') as f:
        json.dump({
            'genres': genres,
            'genre_to_id': genre_to_id
        }, f)

    # Prepare dataset dictionary
    dataset_dict = {
        'audio': [],         # Path to audio file
        'genre': [],         # Genre label
        'track_id': [],      # Track ID
        'title': [],         # Track title
        'artist': []         # Artist name
    }

    # Collect data
    for track_id in tracks.index:
        try:
            # Get audio path
            tid_str = '{:06d}'.format(track_id)
            audio_path = os.path.join(data_dir, 'fma_small',
                                    tid_str[:3], tid_str + '.mp3')

            # Check if file exists
            if not os.path.exists(audio_path):
                continue

            # Get metadata
            genre = tracks.loc[track_id, ('track', 'genre_top')]
            title = tracks.loc[track_id, ('track', 'title')]
            artist = tracks.loc[track_id, ('artist', 'name')]

            # Add to dataset
            dataset_dict['audio'].append(audio_path)
            dataset_dict['genre'].append(genre)
            dataset_dict['track_id'].append(track_id)
            dataset_dict['title'].append(title)
            dataset_dict['artist'].append(artist)

        except Exception as e:
            print(f"Error processing track {track_id}: {e}")
            continue

    # Create Hugging Face dataset
    features = Features({
        'audio': Audio(sampling_rate=16000),  # Will automatically resample to 16kHz
        'genre': ClassLabel(names=genres),
        'track_id': Value('int64'),
        'title': Value('string'),
        'artist': Value('string')
    })

    # Split into train/validation sets
    train_size = int(0.8 * len(dataset_dict['audio']))
    indices = np.random.permutation(len(dataset_dict['audio']))
    train_indices = indices[:train_size]
    val_indices = indices[train_size:]

    # Create train dataset
    train_dict = {
        k: [dataset_dict[k][i] for i in train_indices]
        for k in dataset_dict.keys()
    }
    train_dataset = Dataset.from_dict(train_dict, features=features)

    # Create validation dataset
    val_dict = {
        k: [dataset_dict[k][i] for i in val_indices]
        for k in dataset_dict.keys()
    }
    val_dataset = Dataset.from_dict(val_dict, features=features)

    # Combine into DatasetDict
    dataset_dict = DatasetDict({
        'train': train_dataset,
        'validation': val_dataset
    })

    return dataset_dict

def upload_dataset(dataset_dict, repo_name):
    """Upload dataset to Hugging Face."""
    try:
        # Push to hub
        dataset_dict.push_to_hub(
            repo_name,
            private=True  # Set to False if you want it public
        )

        # Upload genre mapping
        api = HfApi()
        api.upload_file(
            path_or_fileobj="genre_mapping.json",
            path_in_repo="genre_mapping.json",
            repo_id=repo_name,
            repo_type="dataset"
        )

        print(f"Successfully uploaded dataset to {repo_name}")

    except Exception as e:
        print(f"Error uploading dataset: {e}")

````

````python
# Set your Hugging Face repo name
REPO_NAME = "rpmon/fma-genre-classification"

# Prepare dataset
print("Preparing dataset...")
dataset_dict = prepare_fma_dataset("fma_dataset")

# Upload to Hugging Face
print("Uploading to Hugging Face...")
upload_dataset(dataset_dict, REPO_NAME)
````

````python
from transformers import ASTForAudioClassification, AutoFeatureExtractor
from huggingface_hub import HfApi
import torch
import json

def prepare_model_for_upload(model_path, genre_mapping_path):
    """Prepare AST model for upload."""
    # Load genre mapping
    with open(genre_mapping_path, 'r') as f:
        genre_mapping = json.load(f)

    # Load model
    model = ASTForAudioClassification.from_pretrained(
        "MIT/ast-finetuned-audioset-10-10-0.4593",
        num_labels=len(genre_mapping['genres']),
        ignore_mismatched_sizes=True
    )

    # Load trained weights
    model.load_state_dict(torch.load(model_path))

    # Get feature extractor
    feature_extractor = AutoFeatureExtractor.from_pretrained(
        "MIT/ast-finetuned-audioset-10-10-0.4593"
    )

    return model, feature_extractor, genre_mapping

def upload_model(model, feature_extractor, genre_mapping, repo_name):
    """Upload model, feature extractor, and config to Hugging Face."""
    try:
        # Create model card content
        model_card = f"""
# FMA Genre Classification Model

This model is fine-tuned on the FMA (Free Music Archive) dataset for music genre classification.

## Model Description

- Model Type: AST (Audio Spectrogram Transformer)
- Base Model: MIT/ast-finetuned-audioset-10-10-0.4593
- Number of genres: {len(genre_mapping['genres'])}
- Input: Audio waveform (resampled to 16kHz)
- Output: Genre classification

## Genres
{genre_mapping['genres']}

## Usage

```python
from transformers import pipeline

classifier = pipeline(
    "audio-classification",
    model="{repo_name}",
    sampling_rate=16000
)

prediction = classifier("path/to/audio.mp3")
print(prediction)
```

## Training
- Dataset: FMA Small (8,000 tracks)
- Audio Length: 30 seconds
- Training Split: 80%
- Validation Split: 20%
"""

        # Save model card
        with open("README.md", "w") as f:
            f.write(model_card)

        # Push model, feature extractor, and configs to hub
        model.push_to_hub(
            repo_name,
            use_auth_token=True,
            private=True  # Set to False if you want it public
        )

        feature_extractor.push_to_hub(
            repo_name,
            use_auth_token=True
        )

        # Upload genre mapping
        api = HfApi()
        api.upload_file(
            path_or_fileobj="genre_mapping.json",
            path_in_repo="genre_mapping.json",
            repo_id=repo_name,
            repo_type="model"
        )

        print(f"Successfully uploaded model to {repo_name}")

    except Exception as e:
        print(f"Error uploading model: {e}")
````

````python
MODEL_PATH = "best_ast_model.pth"
GENRE_MAPPING_PATH = "genre_mapping.json"
REPO_NAME = "rpmon/fma-genre-classifier"

# Prepare model
print("Preparing model...")
model, feature_extractor, genre_mapping = prepare_model_for_upload(
    MODEL_PATH, GENRE_MAPPING_PATH
)

# Upload to Hugging Face
print("Uploading to Hugging Face...")
upload_model(model, feature_extractor, genre_mapping, REPO_NAME)
````
